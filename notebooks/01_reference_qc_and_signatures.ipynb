# 01 — Reference QC and cell-type signatures

This notebook loads a mock single-cell reference, performs simple QC-style inspection, and derives average expression signatures for each cell type. The workflow is intended to mimic a lightweight reference preparation step for downstream spatial mapping.

```python
from pathlib import Path
import sys

import pandas as pd

sys.path.append(str(Path("..").resolve()))

from src.analysis import (
    GENE_COLUMNS,
    compute_celltype_signatures,
    load_reference,
)
```

```python
reference_df = load_reference("../data/raw/sc_reference.csv")
reference_df.head()
```

```python
print(reference_df.shape)
print(reference_df.columns.tolist())
```

```python
reference_df["cell_type"].value_counts()
```

```python
signatures_df = compute_celltype_signatures(reference_df)
signatures_df.style.format(precision=2).set_caption("Cell-type expression signatures")
```

```python
Path("../data/output").mkdir(parents=True, exist_ok=True)
signatures_df.to_csv("../data/output/celltype_signatures.csv", index=False)
```

The reference data shows distinct average expression patterns across the four mock cell types. These signatures serve as a simplified proxy for reference profiles that can be used to score spatial transcriptomic spots in the next stage of the analysis.
